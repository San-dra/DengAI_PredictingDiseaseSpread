### 1. Loading Datasets

```python
training_label = pd.read_csv("data/dengue_labels_train.csv")
training_features = pd.read_csv("data/dengue_features_train.csv")
test = pd.read_csv("data/dengue_features_test.csv")

training_label['id']=training_label['city'].astype(str) + training_label['year'].astype(str) + training_label['weekofyear'].astype(str) 
training_features['id']=training_features['city'].astype(str) + training_features['year'].astype(str) + training_features['weekofyear'].astype(str)
test['id']=test['city'].astype(str) + test['year'].astype(str) + test['weekofyear'].astype(str)

train = training_features.merge(training_label[['id','total_cases']], on = 'id')
test['total_cases']=-1
display(train.shape,test.shape)
display(train.head())
display(test.head())
```

```python
dataset=train.append(test)
display(dataset.shape)
display(dataset.head())
```

### 2.1. Data Types

```python
dataset.info()
```

```python
dataset['week_start_date']=pd.to_datetime(dataset['week_start_date'])
dataset['week_day_name'] = dataset['week_start_date'].dt.day.astype('int')
dataset["month"] = dataset["week_start_date"].dt.month.astype('int')
```

```python
#Changing datatypes from Objects to Categories.
for col in dataset:
    if dataset.dtypes[col].name == 'object':
        dataset[col] = dataset[col].astype('category')
```
